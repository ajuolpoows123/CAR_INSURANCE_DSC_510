```python
import pdfplumber

# Define the path to your PDF file
pdf_path = r"C:\Users\jamar\Downloads\DSC_510_Car_Insurance_Premium_Data.pdf"

# Open and read the PDF file
with pdfplumber.open(pdf_path) as pdf:
    # Loop through each page
    for page_num, page in enumerate(pdf.pages, start=1):
        text = page.extract_text()
        print(f"Page {page_num} Content:")
        print(text)
        print("\n" + "-"*50 + "\n")  # Separator between pages
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data based on the PDF content you provided
data = {
    "Age Group": ["16-19", "20-25", "26-32", "32-38", "39-44", "45-52", "52-60", "60-68"],
    "Average Miles per Year": [15000, 18000, 18000, 19827, 19987, 14000, 13000, 9000],
    "Most Popular Car Type": [
        "Toyota Corolla", "Tesla", "Mercedes Benz", "Escalade", 
        "Honda Accord", "Jaguar", "Dodge Ram", "Ford Fusion"
    ]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("\nDataFrame:")
print(df)

# Plotting Average Miles per Year by Age Group
plt.figure(figsize=(10, 6))
sns.barplot(x="Average Miles per Year", y="Age Group", data=df, palette="viridis")
plt.title('Average Miles per Year by Age Group', fontsize=16)
plt.xlabel('Average Miles per Year', fontsize=12)
plt.ylabel('Age Group', fontsize=12)
plt.show()
```
